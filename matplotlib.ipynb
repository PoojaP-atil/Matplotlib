```python
y=np.array([35,25,25,15])
lbl=['Apple','Banana','Orange','Grapes']
myexplod=[0.06,0,0,0]
mycolor=['red','yellow','orange','green']
mylabel=[]
plt.pie(y,labels=lbl,startangle=90,explode=myexplod,shadow=True,colors=mycolor)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
tips=sns.load_dataset('tips')
iris=sns.load_dataset('iris')
```

```python
tips
```

```python
a=tips.size
b=tips.shape
c=tips.ndim
d=tips.describe()
e=tips.head()
f=tips.tail()
g=tips.sample()
print('Size',a)
print('Shape',b)
print('Dim',c)
print('Head',e)
print('Tail',f)
print('Describe',d)
print('Sample',g)
```
